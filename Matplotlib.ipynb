linspace( ) 함수는 파이썬의 numpy 모듈에 포함된 함수로서 1차원의 배열 만들기, 그래프 그리기에서 수평축의 간격 만들기 등에 매우 편리하게 사용할 수 있는 함수입니다. 이름에서 알 수 있듯이 Linearly Spaced의 줄임말인 것으로 생각되며, 이 시리즈의 [P026]편에서 간단히 언급한 적이 있는 함수입니다.

사용법은 numpy 모듈을 importing 한 후에 x=np.linspace(start, stop, num)과 같이 적으면 됩니다. start는 배열의 시작값, stop은 배열의 끝값이고, num은 start와 stop 사이를 몇 개의 일정한 간격으로 요소를 만들 것인지를 나타내는 것입니다. 만일 num을 생략하면 디폴트(Default)로 50개의 수열, 즉 1차원 배열을 만들어줍니다.

```python
x = np.linspace(0,10,100) # 0과 10 사이에 100개의 데이터가 있는 배열을 만든다.
```

linspace() 함수는 파이썬의 NumPy에 포함된 함수로 1차원의 배열 만들기 혹은 그래프 그리기에서 수평축의 간격 만들기 등에 매우 편리하게 사용할 수 있는 함수 있다. 사용방법은 x = np(staar, stop, num)과 같은 형태로 사용된다. 이때 num 파라메터를 생략하게 되면 기본적으로 50개의 수열을 생성해 준다.

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## 선 색상과 스타일

```python
plt.plot(x, np.sin(x-0), color = 'blue')          # 색상(color)을 이름으로 지정
plt.plot(x, np.sin(x-1), color = 'g')             # 짧은 색상 코드 예) 'g' = 'green', 'b' = 'blue', 'k' = 'black'
plt.plot(x, np.sin(x-2), color = '0.5')           # 0 ~ 1 사이의 회색조를 지정
plt.plot(x, np.sin(x-3), color = '#FFDD44')       # 16진수로 표현한 RGB 예) #FF(Red값), DD(Green값), 44(Blue값)
plt.plot(x, np.sin(x-4), color = (1.0, 0.2, 0.3)) # RGB 값을 튜플로 표현
plt.plot(x, np.sin(x-5), color = 'chartreuse')    # HTML에서 정해진 색상 이름
```

```python
plt.plot(x, x+0, linestyle = 'solid')
plt.plot(x, x+1, linestyle = 'dashed')
plt.plot(x, x+2, linestyle = 'dashdot')
plt.plot(x, x+3, linestyle = 'dotted')

plt.plot(x, x+4, linestyle = '-')
plt.plot(x, x+5, linestyle = '--')
plt.plot(x, x+6, linestyle = '-.')
plt.plot(x, x+7, linestyle = ':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

## 플롯 조정하기: 축 경계

축 경계를 세밀하게 제어 할 때 사용할 수 있는 함수가 xlim(), ylim()이 있다.

```python
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

## 플롯에 레이블 붙이기

제목과 x, y 축에 레이블 그리고 범례를 붙이는 방법을 살펴본다.

```python
plt.plot(x, np.sin(x))
plt.title('A sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)') # 여기서 label은 그래프의 선에 대한 label을 지정
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')
plt.legend()   # legend = 각주
```

#plt.scatter를 이용한 산점도의 표현

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100) # randn 은 -(마이너스) 좌표도 같이 생성한다.
color = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c = color,
           s = sizes,
           alpha = 0.3,  # alpha 는 투명도 설정
           cmap = 'viridis' ) 

plt.colorbar()
```

```python
from sklearn.datasets import load_iris #scikik-learn
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1],
            s = 100 * features[3],  # s 는 'sizes' 약어
            alpha = 0.2, c = iris.target,   # c 는 'color' 약어
            cmap = 'viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

# 오차 시각화 하기

```python
x= np.linspace(0,10,50)
dy = 0.8  # dy 는 오차범위
y  = np.sin(x) + dy * np.random.randn(50)
```

```python
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3) #ecolor = 에러라인 좌표 색, elinewidth = 에러라인 선의 색
```
